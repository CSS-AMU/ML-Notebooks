```python
x = pdata['Sales']
np.mean(x)

t,p = stats.ttest_1samp(x,7)
print('t-Statistic = %.4f, p-value = %.4f' % (t,p))

(ll,ul) = stats.t.interval(alpha=0.95,df=len(x)-1, loc=np.mean(x), scale=stats.sem(x))
print("CI for Sales = (%.4f,%.4f)" % (ll,ul))
```

```python
y = pdata['Price']
np.mean(y)

t,p = stats.ttest_1samp(y,115)
print('t-Statistic = %.4f, p-value = %.4f' % (t,p))

(ll,ul) = stats.t.interval(alpha=0.95,df=len(y)-1, loc=np.mean(y), scale=stats.sem(y))
print("CI for Price = (%.4f,%.4f)" % (ll,ul))
```

```python
X = pdata['Price']
Y = pdata['Sales']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
print(model.summary())
```

```python
fig1,ax1 = plt.subplots()
sns.regplot('Price','Sales', data = pdata, ax= ax1, ci = False, marker = '.')
ax1.plot([pdata['Price'],pdata['Price']],[pdata['Sales'],model.predict()], linestyle='--',color = 'black')
plt.show()
```

```python
Y_pred = model.predict(X)
residuals = Y - Y_pred
sd_red=np.std(residuals)
a=[-3,-2,-1,0,1,2,3]
b=['r','g','b','k','b','g','r']
sns.scatterplot(Y_pred, residuals)
plt.xlabel('Predicted Sales')
plt.ylabel('Residuals')
for i,j in zip(a,b):
 plt.axhline(i*sd_red,color=j)
```

```python
plt.hist(residuals, color = 'white', edgecolor ='black') #by default bin is set to 10
plt.xlabel('Residuals')
plt.ylabel('Count')
plt.title('Histogram of residuals')
plt.show()
```

```python
sns.distplot(residuals)
```

```python
stats.probplot(residuals, dist="norm", plot=plt,fit = True)
plt.title('Normal Q-Q plot of Residuals')
plt.show()
```

```python
rmse = sk.mean_squared_error(Y, Y_pred, squared= False)
print('RMSE' + ' = ' + str(rmse.round(5)))
```

```python
r2 = sk.r2_score(Y, Y_pred)
print('R2' + ' = ' + str(r2.round(5)))
```

```python
data = pd.get_dummies(pdata, drop_first =True)
data.head()
```

```python
Y = data['Sales']
X = data.drop('Sales', axis=1)

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X, Y)
rfe = RFE(lm,n_features_to_select=8,verbose=1)
rfe = rfe.fit(X,Y)
```

```python
list(zip(X.columns,rfe.support_,rfe.ranking_))
```

```python
X.columns[rfe.support_]
```

```python
X_final = X[X.columns[rfe.support_]]
lmbest = sm.OLS(Y,sm.add_constant(X_final)).fit()
lmbest.summary()
```

```python
pred = lmbest.predict()
resid = Y - pred
```

```python
stats.probplot(resid, dist = "norm", plot=plt)
plt.title('Q-Q plot of residuals')
plt.show()
```

```python
sns.scatterplot(pred, Y, color = 'black')
plt.xlabel('Fitted Sales')
plt.ylabel('Sales')
plt.plot(Y, Y)
plt.title('Observed vs Fitted')
plt.show()
```

```python
sns.scatterplot(pred, resid, color='black')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.axhline(0)
plt.title('Residuals vs Fitted')
plt.show()
```
